```python
import pandas as pd
from sklearn.datasets import fetch_openml
```

```python
data = fetch_openml('titanic', version=1, as_frame=True)
```

```python
df = data.frame
```

```python
df.shape[0]
#Nbr lines
```

```python
df.head()
```

```python
#age
#Transform 'nan' to '-1'
df['age'] = df['age'].fillna(-1)
#Clean the ages, wrong values like '0.98' will turn into 0
df['age'] = df['age'].astype(str).str.split('.').str[0].astype(int)
df.head()
```

```python
filtered_df = df[(df['age'] != 0) & (df['age'] != -1)]
average_age = filtered_df['age'].mean()
median_age = filtered_df['age'].median()
print("Average age :", average_age)
print("Median age :", median_age)

nan_boat_df = df[df['boat'].isna() & (df['age'] != 0) & (df['age'] != -1)]
average_age_nan_boats = nan_boat_df['age'].mean()
median_age_nan_boats = nan_boat_df['age'].median()
print("\nAverage age of the victims:", average_age_nan_boats)
print("Median age of the victims:", median_age_nan_boats)

not_nan_boat_df = df[~df['boat'].isna() & (df['age'] != 0) & (df['age'] != -1)]
average_age_not_nan_boats = not_nan_boat_df['age'].mean()
median_age_not_nan_boats = not_nan_boat_df['age'].median()
print("\nAverage age of survivors:", average_age_not_nan_boats)
print("Median age of survivors:", median_age_not_nan_boats)
```

```python
filtered_df = df[(df['pclass'] != 0) & (df['pclass'] != -1)]
average_pclass = filtered_df['pclass'].mean()
median_pclass = filtered_df['pclass'].median()
print("Average passenger class:", average_pclass)
print("Median passenger class:", median_pclass)

nan_boat_df = df[df['boat'].isna() & (df['pclass'] != 0) & (df['pclass'] != -1)]
average_pclass_nan_boats = nan_boat_df['pclass'].mean()
median_pclass_nan_boats = nan_boat_df['pclass'].median()
print("\nAverage passenger class of the victims:", average_pclass_nan_boats)
print("Median passenger class of the victims:", median_pclass_nan_boats)

not_nan_boat_df = df[~df['boat'].isna() & (df['pclass'] != 0) & (df['pclass'] != -1)]
average_pclass_not_nan_boats = not_nan_boat_df['pclass'].mean()
median_pclass_not_nan_boats = not_nan_boat_df['pclass'].median()
print("\nAverage passenger class of survivors:", average_pclass_not_nan_boats)
print("Median passenger class of survivors:", median_pclass_not_nan_boats)
```
